```python
import requests
from bs4 import BeautifulSoup
web='https://valeriemoreau.com/'
pagina=requests.get(web)
#print(pagina.content)
soup=BeautifulSoup(pagina.content,'html.parser')
resultado=soup.find_all('div',class_='st-content')
print(soup)
print(resultado)
```

```python
import requests
from bs4 import BeautifulSoup
web='https://valeriemoreau.com/'
pagina=requests.get(web)
#print(pagina.content)
soup=BeautifulSoup(pagina.content,'html.parser')
resultado=soup.find_all('div',class_='st-content')
print(soup)
print(resultado)
```
